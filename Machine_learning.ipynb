## First ML


```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
df.head
```

## data preparation


###seprate x and y values

```python
y=df["logS"]
```

```python
x=df.drop("logS", axis=1)
x
```

###splitting the data to train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=100)
X_train
```

##model building

###linear regression

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

#####predict with model

```python
y_lr_train_pred=lr.predict(X_train)
y_lr_test_pred=lr.predict(X_test)
y_lr_test_pred
```

####model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
lr_train_per=mean_squared_error(Y_train,y_lr_train_pred)
lr_train_r2=r2_score(Y_train,y_lr_train_pred)
lr_test_per=mean_squared_error(Y_test,y_lr_test_pred)
lr_test_r2=r2_score(Y_test,y_lr_test_pred)
lr_train_per
lr_train_r2
```

#####table for linear regression results

```python
lr_dt_frame=pd.DataFrame(["Linearregression", lr_test_per,lr_test_r2]).transpose()
lr_dt_frame.columns=["Method","MSE","Rootmeansquare"]
lr_dt_frame
```

###Random forest

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(X_train,Y_train)
```

####predict with model

```python
y_rf_train_pred=rf.predict(X_train)
y_rf_test_pred=rf.predict(X_test)
```

####model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_per=mean_squared_error(Y_train,y_rf_train_pred)
rf_train_r2=r2_score(Y_train,y_rf_train_pred)
rf_test_per=mean_squared_error(Y_test,y_rf_test_pred)
rf_test_r2=r2_score(Y_test,y_rf_test_pred)
print("mse",rf_train_per)
print("root",rf_train_r2)
```

#####table for random forrest

```python
rf_dt_frame=pd.DataFrame(["Randomforrest", rf_test_per,rf_test_r2]).transpose()
rf_dt_frame.columns=["Method","MSE","Rootmeansquare"]
rf_dt_frame
```

##comparison of both methods in table

```python
ml_compare=pd.concat([lr_dt_frame, rf_dt_frame], axis=0)


ml_compare
```
